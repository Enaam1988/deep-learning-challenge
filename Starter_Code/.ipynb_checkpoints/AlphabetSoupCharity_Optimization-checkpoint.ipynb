```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN'.
application_df = application_df.drop(['EIN'], axis=1)
```

```python
# Determine the number of unique values in each column
unique_counts = application_df.nunique()
unique_counts
```

```python
# Checking if I Can drop the STATUS column?
application_df['STATUS'].value_counts()
```

```python
#Checking if I Can drop the SPECIAL_CONSIDERATIONS column?
application_df['SPECIAL_CONSIDERATIONS'].value_counts()
```

```python
#Drop the STATUS and SPECIAL_CONSIDERATIONS
application_df= application_df.drop(columns=['STATUS', 'SPECIAL_CONSIDERATIONS'])
application_df.head()
```

```python
# Look at NAME value counts for binning
name_counts = application_df['NAME'].value_counts()
print(name_counts)
```

```python
# Define the threshold for counts
threshold = 10

# Filter values with counts less than or equal to the threshold
names_to_replace = list(name_counts[name_counts <= threshold].index)

# Replace in the original DataFrame
for name in names_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(name, 'Other')

# Check to make sure binning was successful
print(application_df['NAME'].value_counts())
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
# Create a list of application types to be replaced
application_types_to_replace = application_type_counts[application_type_counts < cutoff_value].index.tolist()
# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value_cls =1000
classifications_to_replace = classification_counts[classification_counts < cutoff_value_cls].index.tolist()
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Identify categorical variables
categorical_variables = application_df.select_dtypes(include=['object']).columns.tolist()
print("Categorical variables:", categorical_variables)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_encoded = pd.get_dummies(application_df)
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_encoded.drop(["IS_SUCCESSFUL"], axis='columns').values
y = application_df_encoded['IS_SUCCESSFUL'].values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
model = tf.keras.models.Sequential()

#Add the first hidden layer
model.add(tf.keras.layers.Dense(units=150, activation='relu', input_dim=number_input_features))

# Add the second hidden layer
model.add(tf.keras.layers.Dense(units=30, activation='sigmoid'))

# Add the third hidden layer
model.add(tf.keras.layers.Dense(units=10, activation='sigmoid'))

# Add the output layer
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the model summary
model.summary()
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = model.fit(X_train_scaled, y_train, epochs=150)
```

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

```python
filename = 'AlphabetSoupCharity_Optimization.h5'

# Save the model to the HDF5 file
model.save(filename)

print(f"Model saved to {filename}.")
```
